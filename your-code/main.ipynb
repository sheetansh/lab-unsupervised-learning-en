```python
print("\n--- STEP 12: OUTLIER DETECTION (IQR Method) ---")
print("Formula: Outliers = values < Q1-1.5*IQR or > Q3+1.5*IQR\n")

outlier_summary = []
for col in dataset_feat:
    Q1 = dataset[col].quantile(0.25)
    Q3 = dataset[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Count outliers
    outliers = dataset[(dataset[col] < lower_bound) | (dataset[col] > upper_bound)]
    n_outliers = len(outliers)
    pct_outliers = (n_outliers / len(dataset)) * 100

    outlier_summary.append({
        'Feature': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower': lower_bound,
        'Upper': upper_bound,
        'Outliers': n_outliers,
        'Pct': f"{pct_outliers:.2f}%"
    })

outlier_df = pd.DataFrame(outlier_summary)
print(outlier_df.to_string(index=False))
```

```python
print("\n--- STEP 13: PARETO PRINCIPLE ANALYSIS ---")
print("Question: Do 20% of customers account for 80% of sales?\n")

# Calculate total spending per customer
customers_pareto = dataset.copy()
customers_pareto['Total_Spending'] = customers_pareto[dataset_feat].sum(axis=1)

# Sort by spending (descending)
customers_pareto = customers_pareto.sort_values('Total_Spending', ascending=False).reset_index(drop=True)

# Calculate cumulative percentage
customers_pareto['Cumulative'] = customers_pareto['Total_Spending'].cumsum()
total_spending = customers_pareto['Total_Spending'].sum()
customers_pareto['Cumulative_Pct'] = (customers_pareto['Cumulative'] / total_spending) * 100

# Find how many customers account for 80% of sales
idx_80 = (customers_pareto['Cumulative_Pct'] >= 80).idxmax()
customers_for_80 = idx_80 + 1
pct_customers_for_80 = (customers_for_80 / len(customers)) * 100

print(f"Total customers: {len(dataset)}")
print(f"Customers for 80% of sales: {customers_for_80}")
print(f"Percentage: {pct_customers_for_80:.2f}%")
print(f"Expected (Pareto): ~20%")
print(f"Result: {'✓ Pareto holds' if pct_customers_for_80 <= 25 else '✗ Pareto does NOT hold'}")

# Plot Pareto curve
plt.figure(figsize=(12, 6))
plt.plot(range(len(customers_pareto)), customers_pareto['Cumulative_Pct'], linewidth=2)
plt.axhline(y=80, color='red', linestyle='--', linewidth=2, label='80% of Sales')
plt.axvline(x=customers_for_80, color='green', linestyle='--', linewidth=2, 
            label=f'{pct_customers_for_80:.1f}% of Customers')
plt.xlabel('Customer Rank (sorted by spending)', fontsize=12)
plt.ylabel('Cumulative Sales (%)', fontsize=12)
plt.title('Pareto Principle Analysis', fontsize=14, fontweight='bold')
plt.legend(fontsize=11)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Your code here
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots

customers = pd.read_csv(path)
spend_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# ── Step 1: Categorical conversion with validation ────────────────
cat_columns = {'Channel': {1: 'HoReCa', 2: 'Retail'},
               'Region':  {1: 'Lisbon',  2: 'Oporto', 3: 'Other'}}

for col, mapping in cat_columns.items():
    # Validate: ensure no unexpected codes exist before mapping
    unexpected = set(customers[col].unique()) - set(mapping.keys())
    assert not unexpected, f"Unexpected values in {col}: {unexpected}"
    customers[col] = customers[col].map(mapping).astype('category')

print("Dtypes after conversion:\n", customers[['Channel','Region']].dtypes)

# ── Step 2: Programmatic missing value audit ──────────────────────
null_report = customers.isnull().sum()
null_pct    = (null_report / len(customers) * 100).round(2)
audit = pd.DataFrame({'null_count': null_report, 'null_%': null_pct})
print("\nMissing Value Audit:\n", audit[audit['null_count'] > 0]
      if audit['null_count'].sum() > 0 else "✔ No missing values.")

# ── Step 3: Outlier removal with full audit trail ─────────────────
# We record how many rows each column's filter removes
# This helps understand which product categories have the most extreme values
audit_trail = {}
customers_clean = customers.copy()

for col in spend_cols:
    Q1, Q3 = customers_clean[col].quantile([0.25, 0.75])
    IQR    = Q3 - Q1
    lower, upper = Q1 - 1.5 * IQR, Q3 + 1.5 * IQR

    before = len(customers_clean)
    # Filter: keep only rows within Tukey's fences
    customers_clean = customers_clean[
        customers_clean[col].between(lower, upper)
    ]
    after = len(customers_clean)
    audit_trail[col] = before - after  # rows removed by this column's filter

print("\nOutlier Removal Audit (rows removed per column):")
print(pd.Series(audit_trail).sort_values(ascending=False))
print(f"\nFinal shape: {customers_clean.shape}  "
      f"({len(customers) - len(customers_clean)} total rows removed)")

# ── Step 4: Skewness before vs after — quantified improvement ─────
skew_before = customers[spend_cols].skew()
skew_after  = customers_clean[spend_cols].skew()

skew_report = pd.DataFrame({
    'Before': skew_before.round(2),
    'After':  skew_after.round(2),
    'Improved': (skew_before - skew_after).round(2)  # positive = better
})
print("\nSkewness Report:\n", skew_report)

# ── Step 5: Log transform preview (for reference before Challenge 3)
# np.log1p(x) = log(x+1) — handles 0 values; reduces right-skewed tails
log_skew = np.log1p(customers_clean[spend_cols]).skew()
print("\nSkewness AFTER log1p transform (preview):\n", log_skew.round(2))
# Values close to 0 = near-normal distribution — ideal for scaling

# ── Step 6: Export cleaned data for next challenges ───────────────
# Always checkpoint your cleaned data before the next pipeline step
customers_clean.to_csv('customers_clean.csv', index=False)
print("\n✔ Cleaned dataset saved to 'customers_clean.csv'")
print("Final column dtypes:\n", customers_clean.dtypes)
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler
import joblib

customers = pd.read_csv('customers_clean.csv')
spend_cols = ['Fresh', 'Milk', 'Grocery', 'Frozen', 'Detergents_Paper', 'Delicassen']

# ── Step 1: Understand the math behind StandardScaler ─────────────
# For each feature column c, the formula is:
#     z_i = (x_i - μ_c) / σ_c
# where μ_c = column mean, σ_c = column std deviation
# After this: mean(z) = 0, std(z) = 1 for every column

# ── Step 2: Fit and transform — using only spend columns ──────────
scaler = StandardScaler()

# IMPORTANT: fit() and transform() are kept conceptually separate
# fit()          → scaler LEARNS the mean & std from training data
# transform()    → scaler APPLIES the formula using what it learned
# fit_transform()→ convenience wrapper that does both in one call

scaler.fit(customers[spend_cols])   # learn mean and std
scaled_array = scaler.transform(customers[spend_cols])  # apply scaling

# Wrap result back into a labelled DataFrame
customers_scale = pd.DataFrame(scaled_array, columns=spend_cols,
                               index=customers.index)  # keep original index

# ── Step 3: Validate with a programmatic assertion ────────────────
# After scaling: each column's mean should be ~0 and std should be ~1
assert all(abs(customers_scale.mean()) < 1e-10), "Means are not ~0 after scaling!"
assert all(abs(customers_scale.std() - 1) < 0.01),  "Stds are not ~1 after scaling!"
print("✔ Validation passed: mean ≈ 0, std ≈ 1 for all columns")

# ── Step 4: Inspect what the scaler learned ───────────────────────
# scaler.mean_ stores the mean of each column (used in z = (x - mean)/std)
# scaler.scale_ stores the std deviation of each column
scaler_params = pd.DataFrame({
    'Learned Mean':  scaler.mean_.round(2),
    'Learned Std':   scaler.scale_.round(2)
}, index=spend_cols)
print("\nScaler Parameters (learned from data):\n", scaler_params)

# ── Step 5: Compare 3 scalers to understand trade-offs ───────────
# StandardScaler: z = (x - mean) / std     → sensitive to outliers
# RobustScaler:   z = (x - median) / IQR   → better with outliers
# MinMaxScaler:   z = (x - min) / (max-min)→ squeezes into [0, 1]

# Since we already removed outliers in Challenge 2,
# StandardScaler is the RIGHT choice here
robust_scale = pd.DataFrame(
    RobustScaler().fit_transform(customers[spend_cols]), columns=spend_cols)
minmax_scale = pd.DataFrame(
    MinMaxScaler().fit_transform(customers[spend_cols]), columns=spend_cols)

print("\nStd of each scaler's output (Standard vs Robust vs MinMax):")
comparison = pd.DataFrame({
    'Standard Std': customers_scale.std().round(3),
    'Robust Std':   robust_scale.std().round(3),
    'MinMax Std':   minmax_scale.std().round(3)
})
print(comparison)

# ── Step 6: Persist scaler — critical for production pipelines ────
# You MUST use the SAME scaler when transforming test/new data.
# Fitting a new scaler on new data = data leakage.
joblib.dump(scaler, 'standard_scaler.pkl')

# Also save the scaled dataset as checkpoint for Challenge 4
customers_scale.to_csv('customers_scale.csv', index=False)
print("\n✔ Scaler saved: standard_scaler.pkl")
print("✔ Scaled data saved: customers_scale.csv")
print("✔ Final customers_scale shape:", customers_scale.shape)

```

```python
# Your code here:
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score

customers       = pd.read_csv('customers_clean.csv')
customers_scale = pd.read_csv('customers_scale.csv')
spend_cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

# ── Step 1: Elbow + Silhouette scores together ────────────────────
# Elbow (Inertia) alone isn't enough — use Silhouette to confirm
# Silhouette score: ranges -1 to +1
#   +1 = well-separated clusters
#    0 = overlapping clusters
#   -1 = wrong cluster assignment
inertia     = []
silhouette  = []
k_range     = range(2, 11)  # silhouette needs at least k=2

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(customers_scale)   # fit + assign labels in one step
    inertia.append(km.inertia_)
    silhouette.append(silhouette_score(customers_scale, labels))

# Print both metrics side by side
metrics = pd.DataFrame({
    'k': k_range,
    'Inertia': [round(v,2) for v in inertia],
    'Silhouette': [round(v,4) for v in silhouette]
})
print(metrics.to_string(index=False))
# k=2 should have the highest silhouette score → confirms elbow

# ── Step 2: Fit final K-Means ──────────────────────────────────────
kmeans = KMeans(n_clusters=2, random_state=42, n_init=10)
kmeans.fit(customers_scale)
customers['labels'] = kmeans.labels_

# ── Step 3: Evaluate with Davies-Bouldin Index ────────────────────
# Davies-Bouldin: LOWER is better (0 = perfect separation)
db_score = davies_bouldin_score(customers_scale, kmeans.labels_)
sil_score = silhouette_score(customers_scale, kmeans.labels_)
print(f"\nSilhouette Score (k=2):     {sil_score:.4f}  ← higher is better")
print(f"Davies-Bouldin Score (k=2): {db_score:.4f}  ← lower is better")

# ── Step 4: Cluster profiling — full statistical breakdown ─────────
# agg() lets us compute multiple stats at once per cluster
cluster_profile = customers.groupby('labels')[spend_cols].agg(['mean','median','std'])
print("\nDetailed Cluster Profiles:\n", cluster_profile.round(0))

# ── Step 5: Label counts ───────────────────────────────────────────
print("\nCluster Label Counts:")
print(customers['labels'].value_counts())
# Cluster 0: 218 customers (HoReCa — Fresh/Frozen heavy)
# Cluster 1: 100 customers (Retail — Grocery/Milk/Detergents heavy)

# ── Step 6: Inspect cluster centers (in scaled space) ─────────────
# cluster_centers_ gives the centroid of each cluster
# Positive z-score = above average, Negative = below average
centers = pd.DataFrame(kmeans.cluster_centers_, columns=spend_cols)
print("\nCluster Centers (z-scores):\n", centers.round(3))
```

```python
from sklearn.cluster import DBSCAN 

# Your code here
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors

customers       = pd.read_csv('customers_clean.csv')
customers_scale = pd.read_csv('customers_scale.csv')
spend_cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

# ── Step 1: Use k-distance plot to find the best eps value ────────
# The k-distance plot is the scientific way to choose eps:
# - For each point, compute its distance to its k-th nearest neighbor
# - Sort distances ascending and plot them
# - The "elbow" of this plot is the optimal eps value
k = 5  # same as DBSCAN's default min_samples
nbrs = NearestNeighbors(n_neighbors=k).fit(customers_scale)
distances, _ = nbrs.kneighbors(customers_scale)

# distances[:, k-1] = distance to the k-th nearest neighbor per point
kth_distances = np.sort(distances[:, k-1])[::-1]
# The elbow of this sorted curve suggests the best eps
print("Top 10 k-distances (look for the elbow):")
print(np.round(kth_distances[:10], 3))
# The elbow appears around 2.0–2.5 — suggesting eps=0.5 is too tight

# ── Step 2: Fit DBSCAN as required by the notebook (eps=0.5) ──────
dbscan = DBSCAN(eps=0.5, min_samples=5)
customers['labels_DBSCAN'] = dbscan.fit_predict(customers_scale)

# ── Step 3: Full audit of results ─────────────────────────────────
label_counts  = customers['labels_DBSCAN'].value_counts().sort_index()
n_clusters    = (label_counts.index != -1).sum()
n_noise       = label_counts.get(-1, 0)
noise_pct     = n_noise / len(customers) * 100

print(f"\nLabel Counts:\n{label_counts}")
print(f"\nClusters: {n_clusters}")
print(f"Noise:    {n_noise} ({noise_pct:.1f}%)")

# ── Step 4: Silhouette score (exclude noise points) ───────────────
# Silhouette is undefined for noise points — filter them out
non_noise_mask = customers['labels_DBSCAN'] != -1
if non_noise_mask.sum() > 1 and len(set(customers.loc[non_noise_mask,'labels_DBSCAN'])) > 1:
    sil = silhouette_score(
        customers_scale[non_noise_mask],
        customers.loc[non_noise_mask, 'labels_DBSCAN']
    )
    print(f"\nSilhouette Score (non-noise only): {sil:.4f}")
else:
    print("\nNot enough non-noise points for silhouette score.")

# ── Step 5: Explain what eps=0.5 means in this context ────────────
# After StandardScaling, the data lives in z-score space
# eps=0.5 means two points must have z-scores within 0.5 std deviations
# For 6-dimensional data — this is extremely tight
# That's why 88.7% of points are noise → eps needs to be larger
print("\n📌 Interpretation:")
print("  eps=0.5 is too small for this 6D dataset.")
print("  Most customers are in sparse space → flagged as noise.")
print("  This is a signal, not an error — see Bonus Challenge 3.")

```

```python
# Your code here:
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, DBSCAN
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score

customers       = pd.read_csv('customers_clean.csv')
customers_scale = pd.read_csv('customers_scale.csv')
spend_cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

km = KMeans(n_clusters=2, random_state=42, n_init=10)
customers['labels'] = km.fit_predict(customers_scale)
db = DBSCAN(eps=0.5)
customers['labels_DBSCAN'] = db.fit_predict(customers_scale)

# ── Statistical evaluation of both algorithms ─────────────────────
# Three metrics compared side by side:
# 1. Silhouette:          Higher is better (+1 ideal)
# 2. Davies-Bouldin:      Lower is better  (0 ideal)
# 3. Calinski-Harabasz:   Higher is better (no upper bound)

# K-Means: all 318 points labelled — straightforward
km_sil = silhouette_score(customers_scale, customers['labels'])
km_db  = davies_bouldin_score(customers_scale, customers['labels'])
km_ch  = calinski_harabasz_score(customers_scale, customers['labels'])

# DBSCAN: exclude noise (-1) for fair evaluation
non_noise = customers['labels_DBSCAN'] != -1
db_scale_filtered  = customers_scale[non_noise]
db_labels_filtered = customers.loc[non_noise, 'labels_DBSCAN']

db_sil = silhouette_score(db_scale_filtered, db_labels_filtered)
db_dbi = davies_bouldin_score(db_scale_filtered, db_labels_filtered)
db_ch  = calinski_harabasz_score(db_scale_filtered, db_labels_filtered)

# Print comparison table
metrics = pd.DataFrame({
    'Metric':     ['Silhouette ↑', 'Davies-Bouldin ↓', 'Calinski-Harabasz ↑',
                   'Points Labelled', 'Noise Points'],
    'K-Means':    [round(km_sil,4), round(km_db,4), round(km_ch,2), 318, 0],
    'DBSCAN':     [round(db_sil,4), round(db_dbi,4), round(db_ch,2), 36, 282]
})
print(metrics.to_string(index=False))

# ── GroupBy means — full breakdown ────────────────────────────────
print("\nK-Means — Mean spend per cluster:")
print(customers.groupby('labels')[spend_cols].mean().round(0))

print("\nDBSCAN — Mean spend per label (incl. noise):")
print(customers.groupby('labels_DBSCAN')[spend_cols].mean().round(0))

# ── Verdict ───────────────────────────────────────────────────────
# K-Means wins for this dataset because:
# 1. It assigns labels to ALL customers (100% coverage)
# 2. Its clusters are statistically better (higher Silhouette, lower DB index)
# 3. DBSCAN with eps=0.5 marks 88.7% of data as noise
#    → useless for business insight
# DBSCAN would win on datasets with:
# - Irregular shaped clusters
# - Clear density gaps
# - True noise/outliers that should be excluded
```

# Bonus Challenge 2 - Changing K-Means Number of Clusters

As we mentioned earlier, we don't need to worry about the number of clusters with DBSCAN because it automatically decides that based on the parameters we send to it. But with K-Means, we have to supply the `n_clusters` param (if you don't supply `n_clusters`, the algorithm will use `8` by default). You need to know that the optimal number of clusters differs case by case based on the dataset. K-Means can perform badly if the wrong number of clusters is used.

In advanced machine learning, data scientists try different numbers of clusters and evaluate the results with statistical measures (read [here](https://en.wikipedia.org/wiki/Cluster_analysis#External_evaluation)). We are not using statistical measures today but we'll use our eyes instead. In the cells below, experiment with different number of clusters and visualize with scatter plots. What number of clusters seems to work best for K-Means?

```python
# Your code here:
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

customers       = pd.read_csv('customers_clean.csv')
customers_scale = pd.read_csv('customers_scale.csv')
spend_cols = ['Fresh','Milk','Grocery','Frozen','Detergents_Paper','Delicassen']

# ── Step 1: Collect inertia AND silhouette scores ─────────────────
# Using two metrics is much better than just inertia alone
# Silhouette score: measures how well-separated the clusters are
inertia    = []
silhouette = []
k_range    = range(2, 9)

for k in k_range:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = km.fit_predict(customers_scale)
    inertia.append(km.inertia_)
    # silhouette_score needs the data AND the labels
    silhouette.append(silhouette_score(customers_scale, labels))

# ── Step 2: Plot both metrics side by side ────────────────────────
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

ax1.plot(list(k_range), inertia, 'bo-')
ax1.axvline(x=2, color='red', linestyle='--', label='Best k=2')
ax1.set_title("Inertia (lower = better)")
ax1.set_xlabel("k"); ax1.set_ylabel("Inertia")
ax1.legend()

ax2.plot(list(k_range), silhouette, 'go-')
ax2.axvline(x=2, color='red', linestyle='--', label='Best k=2')
ax2.set_title("Silhouette Score (higher = better)")
ax2.set_xlabel("k"); ax2.set_ylabel("Silhouette Score")
ax2.legend()

plt.suptitle("K-Means: Metrics vs Number of Clusters")
plt.tight_layout()
plt.show()

# ── Step 3: Fit and visualize k=3 and k=5 ────────────────────────
# Show what happens to clusters when we over-split or under-split
for k in [2, 3, 5]:
    km = KMeans(n_clusters=k, random_state=42, n_init=10)
    customers[f'labels_k{k}'] = km.fit_predict(customers_scale)

# Scatter: 3 values of k side by side on Grocery vs Fresh
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
for i, k in enumerate([2, 3, 5]):
    axes[i].scatter(customers['Grocery'], customers['Fresh'],
                    c=customers[f'labels_k{k}'], cmap='tab10', alpha=0.6)
    sil = silhouette_score(customers_scale, customers[f'labels_k{k}'])
    axes[i].set_title(f"k={k}  |  Silhouette={sil:.3f}")
    axes[i].set_xlabel("Grocery"); axes[i].set_ylabel("Fresh")
plt.tight_layout(); plt.show()

# ── Step 4: Print cluster sizes ───────────────────────────────────
for k in [2, 3, 5]:
    print(f"k={k} cluster sizes:", customers[f'labels_k{k}'].value_counts().to_dict())
```

# Bonus Challenge 3 - Changing DBSCAN `eps` and `min_samples`

Experiment changing the `eps` and `min_samples` params for DBSCAN. See how the results differ with scatter plot visualization.

```python
# Your code here
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt

customers       = pd.read_csv('customers_clean.csv')
customers_scale = pd.read_csv('customers_scale.csv')

# ── Step 1: k-Distance Plot — scientific way to choose eps ────────
# For each point, compute the distance to its 5th nearest neighbor
# Sort descending and plot — the "elbow" is the ideal eps
k = 5
nbrs = NearestNeighbors(n_neighbors=k).fit(customers_scale)
distances, _  = nbrs.kneighbors(customers_scale)

# distances[:, k-1] = each point's distance to its k-th nearest neighbor
kth_distances = np.sort(distances[:, k-1])[::-1]

plt.figure(figsize=(8, 4))
plt.plot(kth_distances, linewidth=2)
plt.axhline(y=1.0, color='red', linestyle='--', label='eps ≈ 1.0 (elbow)')
plt.xlabel("Points (sorted)")
plt.ylabel("Distance to 5th Nearest Neighbor")
plt.title("k-Distance Plot — Choosing eps for DBSCAN")
plt.legend()
plt.show()
# The elbow is at ~1.0 — that is your scientifically grounded eps

# ── Step 2: Experiment with 6 configurations ─────────────────────
configs = [
    (0.5,  5), (1.0,  5), (1.0, 3),
    (1.5,  5), (2.0,  5), (2.5, 5)
]

fig, axes = plt.subplots(2, 3, figsize=(15, 9))
for ax, (eps, ms) in zip(axes.flatten(), configs):
    db = DBSCAN(eps=eps, min_samples=ms)
    labels = db.fit_predict(customers_scale)

    n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
    n_noise    = (labels == -1).sum()

    ax.scatter(customers['Detergents_Paper'], customers['Milk'],
               c=labels, cmap='tab10', alpha=0.6, s=20)
    ax.set_title(f"eps={eps}, ms={ms}\n"
                 f"clusters={n_clusters} | noise={n_noise}", fontsize=9)
    ax.set_xlabel("Detergents_Paper")
    ax.set_ylabel("Milk")

plt.suptitle("DBSCAN: Experimenting with eps and min_samples", fontsize=13)
plt.tight_layout()
plt.show()
```
